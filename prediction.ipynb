```python
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
### Load the pickle file
model=load_model('model.h5')
## Load the encoder and scaler
with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)

with open('onehot_encoder_geo.pkl','rb') as file:
    label_encoder_geo=pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)
```

```python
input_data = {
    'CreditScore': 600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
input_df=pd.DataFrame([input_data])
input_df
```

```python
## One hot encoding of geography
geo_df=pd.DataFrame(label_encoder_geo.transform([input_df['Geography']]).toarray(),columns=label_encoder_geo.get_feature_names_out())
geo_df
```

```python
## Enocoding gender
input_df['Gender']=label_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
## for joining columnwise use axis=1
final_df=pd.concat([input_df.drop(columns=['Geography']),geo_df],axis='columns')
```

```python
final_df
```

```python
# scaling the numerical values
scaled_df=scaler.transform(final_df)
scaled_df
```

```python
prediction=model.predict(scaled_df)
prediction
```
